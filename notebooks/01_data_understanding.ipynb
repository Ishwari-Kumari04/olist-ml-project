## Orders Table

### Business Purpose

The Orders table captures the complete lifecycle of every customer order, including purchase, approval, shipment, delivery, and current order status.

---

### Initial Observations

#### Number of Columns
The **Orders** table contains **8 columns**, each representing a different stage in the order lifecycle.

#### Primary Key
- **`order_id`** uniquely identifies each order.

#### Foreign Key
- **`customer_id`** links the Orders table to the **Customers** table.

---

### Timestamp Columns

The Orders table captures the complete lifecycle of an order through multiple timestamps.

| Column | Business Meaning |
|--------|------------------|
| **`order_purchase_timestamp`** | The exact date and time when the customer placed the order. |
| **`order_approved_at`** | The date and time when the payment/order was approved. |
| **`order_delivered_carrier_date`** | The timestamp when the seller handed the package over to the logistics carrier. |
| **`order_delivered_customer_date`** | The actual delivery date and time when the customer received the order. |
| **`order_estimated_delivery_date`** | The promised/estimated delivery date communicated to the customer. |

---

### Order Status

The **`order_status`** column represents the current stage of an order within its lifecycle (e.g., delivered, shipped, canceled, processing).

---

### Feature Engineering Opportunities

The timestamp columns can be transformed into meaningful business features for analytics and machine learning.

| Feature | Calculation | Business Value |
|---------|-------------|----------------|
| **Delivery Time** | `Delivered Date - Purchase Date` | Measures the overall customer delivery experience. |
| **Approval Time** | `Approval Date - Purchase Date` | Measures payment/order approval efficiency. |
| **Seller Processing Time** | `Carrier Pickup Date - Approval Date` | Evaluates seller processing performance. |
| **Shipping Time** | `Customer Delivery Date - Carrier Pickup Date` | Measures logistics/carrier performance. |
| **Delivery Delay** | `Customer Delivery Date - Estimated Delivery Date` | Indicates whether an order was delivered early, on time, or late. |

## Customers Table

### Business Purpose

The Customers table stores customer identifiers and geographical information, enabling customer-level and regional analysis.

---

### Initial Observations

#### Number of Columns
The **Customers** table contains **5 columns** and stores customer identifiers along with their geographical information.

#### Primary Key
- **`customer_id`** uniquely identifies each customer record.

---

### Customer Information

The Customers table contains only location-related attributes and customer identifiers. No personally identifiable information (PII) such as customer names, email addresses, or phone numbers is stored.

| Column | Business Meaning |
|--------|------------------|
| **`customer_zip_code_prefix`** | ZIP code prefix of the customer's location. |
| **`customer_city`** | City where the customer is located. |
| **`customer_state`** | State where the customer is located. |

---

### Customer Identifiers

| Column | Business Meaning |
|--------|------------------|
| **`customer_id`** | Unique identifier for a customer record. Used to link the Customers table with the Orders table. |
| **`customer_unique_id`** | Represents the actual unique customer across multiple purchases. A single customer may have multiple `customer_id` values but only one `customer_unique_id`. *(To be verified during further analysis.)* |

---

### Initial Business Observations

- Customer privacy is maintained as no personally identifiable information (PII) is available.
- Customer location can be analyzed at the city, state, and ZIP code prefix levels.
- The presence of both `customer_id` and `customer_unique_id` suggests that a customer may place multiple orders over time. This hypothesis will be validated in the next phase of the analysis.

---

### Feature Engineering Opportunities

The Customers table primarily contributes **customer-level demographic and geographical features**. Although the table itself is small, these features become valuable when combined with orders, payments, and reviews.

| Feature | Source Column(s) | Business Value |
|---------|------------------|----------------|
| **Customer State** | `customer_state` | Analyze regional purchasing behavior and demand patterns. |
| **Customer City** | `customer_city` | Capture city-level buying trends and customer segmentation. |
| **Customer ZIP Code Prefix** | `customer_zip_code_prefix` | Enable fine-grained geographical analysis and delivery performance evaluation. |
| **Repeat Customer Flag** | `customer_unique_id` | Identify customers who have placed multiple orders. |
| **Customer Order Count** *(Derived)* | Orders + `customer_unique_id` | Measure customer loyalty and purchase frequency. |
| **Customer Lifetime** *(Derived)* | Orders + Purchase Timestamps | Calculate the duration between the customer's first and last purchase. |
| **Average Purchase Interval** *(Derived)* | Orders + Purchase Timestamps | Measure how frequently a customer places orders. |

## Order Items Table

### Business Purpose

The Order Items table records every individual item purchased within an order, linking customers, products, and sellers at the transaction level.

---

### Initial Observations

#### Number of Columns
The **Order Items** table contains **7 columns**. Each row represents a single item within an order, making it a transaction-level table.

#### Primary Key

- **Composite Primary Key:** (`order_id`, `order_item_id`)
- `order_id` identifies an order, while `order_item_id` identifies an individual item within that order. Together, they uniquely identify each transaction record.

#### Foreign Keys

- **`order_id`** → Links to the **Orders** table.
- **`product_id`** → Links to the **Products** table.
- **`seller_id`** → Links to the **Sellers** table.

#### Key Observations

- **`order_id` is not unique**, as a single order can contain multiple products.
- Each row represents **one purchased item**, allowing multiple rows for the same order.
- **`price`** represents the price of an individual item.
- **`freight_value`** represents the shipping cost associated with that item.

---

### Business Questions

The Order Items table enables analysis of transaction-level purchasing behavior and seller performance.

- Which products generate the highest revenue?
- Which sellers sell the most products?
- Which products are frequently purchased together? *(Future recommendation system)*
- Which orders incur the highest shipping cost?
- Which product categories contribute the most to revenue?
- What is the average number of items per order?

---

### Feature Engineering Opportunities

The Order Items table contributes transaction-level features that describe the cost and composition of each order.

| Feature | Source Column(s) | Business Value |
|---------|------------------|----------------|
| **Item Price** | `price` | Represents the price of an individual purchased item. |
| **Freight Cost** | `freight_value` | Captures the shipping cost associated with each item. |
| **Total Item Cost** *(Derived)* | `price + freight_value` | Represents the total cost of purchasing an item, including shipping. |
| **Items per Order** *(Derived)* | `order_id` | Measures basket size by counting the number of items in an order. |
